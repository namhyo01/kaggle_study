```python
# 수익률: 투자한 자본에 대한 수익(혹은 손실)의 비율
# 수익률 = (나중가격 - 처음가격) / 처음가격
# ex1) 하루는 10%이익을, 하루는 10%손해를 봤다면 수익률이 0%일까?
#  1000 => +10%(1100) => -10% (잔액990)
# ex2) 수익과 손실의 순서를 바꾸면?
#  1000 => -10%(900) => +10%(잔액990)

# 일반수익률과 로그수익률
# 일반수익률(R) = (나중가격-처음가격) / 처음 가격 = (P-P0)/P0 = P/P0 - 1
# 로그수익률(R) = ln(나중가격/처음가격) = lnP/P0 = ln(P) - ln(P0)
# (표)
# 거래일,   가격,    일반수익률,   로그수익률
#  1일     1000          .             .
#  2일     1300          30.00%     26.24%
#  3일      800         -38.46%    -48.55%
#  4일     1300          62.50%     48.55$
#수익률합계  39%         38.65%      9.53%
#최종수익률  10%         10.00%      9.53%

# 일반수익률 
#  * 재투자해서 발생하는 손실까지 포함
#  * 각 거래에서 발생하는 손실률의 합과 최종손실률이 달라진다.
# 로그수익률
#  * 최종수익률과 수익률의 합계가 일치
#  * 금융 분야에서는 주로 로그 수익률을 사용.
```
